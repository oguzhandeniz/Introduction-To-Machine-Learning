```python
timesData = pd.read_csv("/kaggle/input/timesdata/timesData.csv")
```

```python
timesData.head()
```

```python
dataframe = timesData[timesData.year == 2015]
```

```python
trace1 = go.Scatter3d(
    x=dataframe.world_rank, 
    y=dataframe.research,   
    z=dataframe.citations,  
    mode='markers',        
    marker=dict(
        size=dataframe.teaching, 
        color='rgb(255,0,0)',   
    )
)

data = [trace1] 

layout = go.Layout(
    margin=dict(  
        l=0,
        r=0,
        b=0,
        t=0       
    )
    
)
fig = go.Figure(data=data, layout=layout) 

iplot(fig)
```
