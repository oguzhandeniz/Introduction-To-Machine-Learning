```python
import matplotlib.pyplot as plt
import numpy as np

# DATA
x=np.arange(0,10,0.1)

# [0, 0.1 , 0.2 , 0.3 ..... 9.9]

y=np.sin(x)

# Grap creation
plt.plot(x,y)

# Determine chart propertioe
plt.title("Sinüs Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#DATA
x = np.random.rand(34)
y = np.random.rand(34)

colors = np.random.rand(34)
sizes = np.random.randint(34,200,34)

plt.scatter(x, y, c = colors ,s = sizes)
plt.title("Nokta Grafiği")
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#DATA
data=np.random.randn(2500)

plt.hist(data,bins =  40)

plt.title("Histogram")
plt.xlabel("Degerler")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#DATA
x = ['A','B','C','D','E']

y = np.random.randint(1,15,5)

plt.bar(x,y)
plt.title("Sinus Grafiği")
plt.xlabel("Kategoriler")
plt.ylabel("Değerler")

plt.show()
```
